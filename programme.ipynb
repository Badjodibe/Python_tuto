<h3 style = "text-decoration-line:underline;"> Aperçu de la formation</h3>
<p> Ce programme concerne la formation en python qui sera sur une durée de deux mois. python est un langae de programmation <br>
multidisciplinaire, il peut être utiliser dans les recherche scientifique, développement informatique, réseau informatique, <br>
, la finance etc, la liste est longue en conséquence dire apprendre python est très vaste étant données son application. Dans cette formation <br>
Dans cette formation, nous allons nous concentrer sur ce qui leurs sont commun,  les bases de la programmation python, découvrir <br> la programmation orientée object. A la fin nous verrons l'application dans  une domaine bien précise qui devient de plus en <br> plus populaire : La science de données, c'est un domaine multidisciplinaire, impliquant la mathématique, programmation, <br> statistique, intéllignece artificielle. On va simplement appliquer cette science sur un problème concrète en utilisant python
</p>
